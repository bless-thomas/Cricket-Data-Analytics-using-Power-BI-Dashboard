**Batting Summary**

```python
with open('batting_summary.json') as f:
    data=json.load(f)
all_records =[] #since for each team1 vs team 2 there's a separate batting summary we want to collect all of that into a single table

for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting= pd.DataFrame(all_records)
```

```python
df_batting.head(11)
```

```python
# converting the ' dismissal' column into out/not out by looking at blank values
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "not_out" if pd.isna(x) else ("out" if len(x)>0 else "not_out"))
```

```python
df_batting
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
```

```python
df_batting.head(11)
```

```python
#code to find what all extra characters are there in the batsman column
import string
unique_chars = set(''.join(df_batting['batsmanName']))
non_alpha_chars = unique_chars - set(string.ascii_letters)
print(non_alpha_chars)
```

```python
df_batting["batsmanName"] = df_batting['batsmanName'].apply(lambda x: x.replace("\xa0",""))
df_batting["batsmanName"] = df_batting['batsmanName'].apply(lambda x: x.replace("€",""))
df_batting["batsmanName"] = df_batting['batsmanName'].apply(lambda x: x.replace("â",""))
```

```python
'''In order the link the table 'match_summary' and 'batting_summary' we need a connecting column, the only common thing they
have is the team names, however the team names need to be in pairs and any order can exist so each pair has two versions.
For this we create a dictionary with all the possible combinations of teams as KEYS and match_id as VALUES to distinguish.'''
```

```python
match_ids_dict= {}

for index, row in df_match.iterrows():
    key1= row["team1"] + " Vs " + row["team2"]
    key2= row["team2"] + " Vs " + row["team1"]
    
    match_ids_dict[key1]=row["match_id"]
    match_ids_dict[key2]=row["match_id"]
```

```python
#Creating a new column in batting_summary dataframe of match_id to get a primary key
```

```python
df_batting["match_id"]=df_batting["match"].map(match_ids_dict)
df_batting
```

```python
df_batting.to_csv("cricket_analysis_batting_summary.csv", index=False)
```

 # Bowling Summary

```python
with open("bowling_summary.json") as f:
    data = json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec["bowlingSummary"])
```

```python
df_bowling=pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head(30)
```

```python
df_bowling["match_id"]=df_bowling["match"].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv("cricket_analysis_bowling_summary.csv", index=False)
```
